```python
import numpy as np
import cvxpy as cp
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2, 5, 1000)
f0 = lambda x : x**2 + 1
f1 = lambda x : (x - 4)*(x-2)
```

```python
fig, ax = plt.subplots()
# ax.plot(x, f0(x), label='$f_{0}(x)$', color='#34495E')
ax.plot(x, f0(x), label='$f_{0}(x)$')
# ax.plot(x, f1(x), label='$f_{1}(x)$', color='#34495E')
ax.plot(x, f1(x), label='$f_{1}(x)$')

ax.axvline(x=2, color='black', linestyle=':', linewidth=1)
ax.axvline(x=4, color='black', linestyle=':', linewidth=1)
ax.axhline(y=0, color='black', linewidth=1)

ax.axvspan(2, 4, color='grey', alpha=0.3)

ax.set_xlabel('x')
ax.legend()
plt.show()
```

```python
L = lambda tup : (1 + tup[1])*(tup[0]**2) - tup[1]*6*tup[0] + 1 + tup[1]*8
```

```python
fig, ax = plt.subplots()
# ax.plot(x, f0(x), label='$f_{0}(x)$', color='#34495E')
ax.plot(x, f0(x), label='$f_{0}(x)$')
# ax.plot(x, f1(x), label='$f_{1}(x)$', color='#34495E')
ax.plot(x, f1(x), label='$f_{1}(x)$')
lambdas = [1, 5, 10, 20]
for l in lambdas:
    ax.plot(x, L((x, l)), label=f'$L(x, \\lambda = {l})$')

ax.axvline(x=2, color='black', linestyle=':', linewidth=1)
ax.axvline(x=4, color='black', linestyle=':', linewidth=1)
ax.axhline(y=0, color='black', linewidth=1)

ax.axvspan(2, 4, color='grey', alpha=0.3)

ax.set_xlabel('x')
ax.set_ylim(-5, 30)

ax.legend()
plt.show()
```

```python
x1 = np.linspace(-10, 10, 1000)
fig, ax = plt.subplots()
val = lambda tup : (1 + tup[1])*(tup[0]**2) - tup[1]*6*tup[0]
for lm in lambdas:
    ax.plot(x1, val((x1, lm)))
```

```python
x1 = np.linspace(-10, 10, 1000)
x1 = [x for x in x1 if x!=0]
x_tilde = lambda lm : 3*lm / (1 + lm)
plt.plot(x1, [x_tilde(x) for x in x1])

plt.show()
```
